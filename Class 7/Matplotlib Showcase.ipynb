```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

### Subplotting

```python
%matplotlib inline
# Subplots can be created using the command plt.subplot(XYz) -- XxY matrix, z th quadrant
dummy_data = np.arange(10)
plt.subplot(221)
plt.scatter(dummy_data,dummy_data**3,cmap=plt.colormaps())
plt.plot(dummy_data,dummy_data**2, c='R')
plt.subplot(222)
plt.hist(dummy_data%.7)
plt.subplot(223)
plt.boxplot(dummy_data)
plt.subplot(224)
plt.barh(dummy_data,dummy_data)
```

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])

df2.plot.area()
df2.plot.density()
df2.plot.scatter('a','c')
```

```python
%matplotlib inline
data = load_digits()
plt.clf()
plt.cla()
plt.gray()
axes = Axes3D(plt.figure())
axes.scatter3D(data['data'][0],data['data'][1],data['data'][2],cmap='jet')
```
